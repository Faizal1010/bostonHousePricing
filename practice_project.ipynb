```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("/content/boston_house_prices.csv")
dataset
```

```python
# Checking for Missing Values
dataset.isnull().sum()
```

```python
#### Exploratory Data Analysis (EDA)
## Correlation
dataset.corr()
```

```python
plt.scatter(dataset['INDUS'], dataset['MEDV'], s= 10)  #Analyzing the relationship between INDUS and MEDV
```

```python
#Independent and Dependent Features

X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
## Standardize the Dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()

regression.fit(X_train, y_train)
```

```python
regression.coef_
```

```python
#Checking the Parameters on which the model has been trained
regression.get_params()
```

```python
#Prediction with Test data
pred = regression.predict(X_test)
```

```python
#Plot a scatter plot for the prediction
plt.scatter(pred, y_test)
```

```python
#calculating the residuals (error)

residuals = y_test-pred
residuals
```

```python
sns.displot(residuals, kind='kde')
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

mean_squared_error(y_test, pred)
```

```python
mean_absolute_error(y_test, pred)
```

```python
# Performance Metrics

from sklearn.metrics import r2_score, ad

r2_score(y_test, pred)
```
